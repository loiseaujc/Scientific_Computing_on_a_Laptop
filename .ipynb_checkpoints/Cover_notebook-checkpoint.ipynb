````python
```
{code-cell} ipyton3
:tags: ["hide-input"]

import numpy as np
import matplotlib.pyplot as plt

def rule_index(triplet):
    L, C, R = triplet
    index = 7 - (4*L + 2*C + R)
    return int(index)

def CA_run(initial_state, nsteps, rule_number=30):
    rule_string = np.binary_repr(rule_number, 8)
    rule = np.array([int(bit) for bit in rule_string])
    
    ncells = len(initial_state)
    state = np.zeros((nsteps, ncells))
    state[0] = initial_state
    
    for i in range(1, nsteps):
        all_triplets = np.stack([
            np.roll(state[i-1], 1), state[i-1], np.roll(state[i-1], -1)
        ])
        
        state[i] = rule[np.apply_along_axis(rule_index, 0, all_triplets)]
        
    return state

n, nsteps = 256, 256
initial_state = np.random.randint(0, 2, n)
state = CA_run(initial_state, nsteps)

fig, ax = plt.subplots(1, 1, figsize=(12, 12))

ax.matshow(state, cmap=plt.cm.binary)
ax.axis("off");
```
````
